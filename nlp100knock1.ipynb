文字列”stressed”の文字を逆に（末尾から先頭に向かって）並べた文字列を得よ．

```python
def q00(arg='stressed'):
    return arg[::-1]
def q00a(arg='stressed'):
    return ''.join(list(reversed(arg)))
print(q00())
print(q00a())
```

「パトカー」＋「タクシー」の文字を先頭から交互に連結して文字列「パタトクカシーー」を得よ．

```python
def q02(arg1='パトカー',arg2='タクシー'):
    for i,j in zip(arg1,arg2):
        print(i,end='')
        print(j,end='')
    print()
q02()
```

“Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.”という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ．

```python
def q03(arg='Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.'):
    text=arg.replace('.','').replace(',','')
    ans=[len(i)for i in text.split()]
    return ans
print(q03())
```

“Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭に2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```python
def q04(arg='Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'):
    words=arg.split()
    dic={}
    firstnum=[1,5,6,7,8,9,15,16,19]
    for i in range(len(words)):#語数分
        for word in words:#語を1語ずつ
            if i in firstnum:
                dic[word[0:2]]=i+1
            else:
                dic[word[0:1]]=i+1
    return dic
print(q04())  
```

与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．

```python
def q05(arg,n):
    text=arg.replace(' ','')
    start=0
    charans=[]
    while True:
        goal=start+n
        if goal>=len(text):
            break
        charans.append(text[start:goal])
        start=goal
    
    words=arg.split()
    #print(words)
    start=0
    wordans=[]
    while True:
        goal=start+n
        if goal>len(words):
            break
        wordans.append(''.join(words[start:goal]))
        start=goal
    return charans,wordans
print(q05('I am an NLPer',2))
```

```python
def charbi(arg):
    start=0
    charans=[]
    while True:
        goal=start+2
        if goal>len(arg):
                break
        charans.append(arg[start:goal])
        start=goal
    return charans
def q06(X='paraparaparadise',Y='paragraph',word='se'):
    Xset=set(charbi(X))
    Yset=set(charbi(Y))
    return Xset|Yset,Xset&Yset,Xset-Yset,word in Xset,word in Yset
q06()
```

与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

英小文字ならば(219 - 文字コード)の文字に置換
その他の文字はそのまま出力
この関数を用い，英語のメッセージを暗号化・復号化せよ．

```python
def q08(arg):
    ans=''
    for word in arg:
        if word.islower():
             ans+= chr(219 - ord(word))
        else:
            ans+=word
    return ans
print(q08('ABii'))
```

スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```python
import random
def q09(arg):
    words=arg.split()
    print(words)
    lst=[]
    for word in words:#1語ごと
        if len(word)>4:
            word=word[0]+''.join(random.sample(word[1:-1], len(word[1:-1])))+word[-1]
            lst.append(word)
        else:
            lst.append(word)
    return ' '.join(lst)
print(q09('I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .'))
```
